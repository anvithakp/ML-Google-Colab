```
#create url variables for each zillow page
with requests.Session() as s:
   
    url1 = 'https://www.zillow.com/san-diego-county-ca/sold/'

    r1 = s.get(url1, headers=req_headers)
    
    url_links = [url1]
```

```

#calculate the zestimate and insert into a dataframe
zillow_zestimate = []
for link in df['links']:
    r = s.get(link, headers=req_headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    home_value = soup.find("div",{'class': "zestimate-value"})
    zillow_zestimate.append(home_value)

cols=['zestimate']
zestimate_result = pd.DataFrame(zillow_zestimate, columns=cols)
```

```
#convert zestimate column to float, and remove , and $
import re
def apply_regex(zestimate_result):
  #  print(zestimate_result2['HOA'])
   p = re.search(">\$(.*?)\<", str(zestimate_result['zestimate']))
   if str(p) in 'None':
     return 0
   else:
     return p.group(1)

#zestimate_result99 = zestimate_result2.copy(True)
zestimate_result['zestimate'] = zestimate_result.apply(apply_regex,axis=1)
```

```
zestimate_result['zestimate']
```

```
def apply_regex_sale(zestimate_result):
  #  print(zestimate_result2['HOA'])
   p = re.search(">\$(.*?)<", str(zestimate_result))
   if str(p) in 'None':
     return 0
   else:
     return p.group(1).replace(',', '')


# links=["https://www.zillow.com/homedetails/615-Fredricks-Ave-UNIT-140-Oceanside-CA-92058/16592979_zpid/"]

listprices=[]
refinances=[]
for link in df['links']:
    scores_page = s.get(link, headers=req_headers)
    soup = BeautifulSoup(scores_page.content, 'html.parser')

    temp2=soup.findAll("p",{'class': "Text-aiai24-0 StyledParagraph-sc-18ze78a-0 hiYlnH"})
    if len(temp2) > 0:
      listprices.append(apply_regex_sale(str(temp2[0])))
    else:
      listprices.append(0)
    if len(temp2) > 2:
      refinances.append(apply_regex_sale(str(temp2[-1])))
    else:
      refinances.append(0)
print(listprices,refinances)


 
```

```
zillow_hoa = []

import re
total = 0
linkCount = 0
for link in df['links']:
    if (linkCount % 50 == 0):
      print(linkCount)
    linkCount += 1
    r = s.get(link, headers=req_headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    hoa = soup.select(".ds-home-fact-list .ds-home-fact-list-item .Text-aiai24-0.gbgvjX")
    # print(hoa)
    found = '0'
    for i in hoa:
      p = re.search("Text-aiai24-0 gbgvjX\">\$(.*?)\/mo", str(i))
      if str(p) in 'None':
       continue
      else:
       found = p.group(1)
    # print(hoa)
    zillow_hoa.append(found)
print(zillow_hoa)   
```

```
cols=['HOA']
zestimate_result3 = pd.DataFrame(zillow_hoa, columns=cols)
```

```
import re

def apply_regex(zestimate_result):
  #  print(zestimate_result2['HOA'])
   p = re.search("(\d+) out of 100", str(zestimate_result))
   if str(p) in 'None':
     return 0
   else:
     return p.group(1)

walk_scores_urls = []
walk_scores = []

for link in df['links']:
    text=link.split('/')
    url="https://www.walkscore.com/score/"+text[4]
    walk_scores_urls.append(url)
# print(walk_scores_urls)
# link='https://www.walkscore.com/score/4095-Bonita-Road--132_rb'
for link in walk_scores_urls:
  scores_page = s.get(link, headers=req_headers)
  soup = BeautifulSoup(scores_page.content, 'html.parser')
  est = soup.find(id='score-description-sentence')
  walk_score=apply_regex(est)
  walk_scores.append(walk_score)

print(walk_scores)

```

```
cols=['Walk_Score']
Z_walk_score = pd.DataFrame(walk_scores, columns=cols)
```

```
def apply_regex_sold(zestimate_result):
  #  print(zestimate_result2['HOA'])
   p = re.search("\$([\d,]+)", str(zestimate_result))
   if str(p) in 'None':
     return 0
   else:
     return p.group(1)

sold_prices = []

# links=["https://www.zillow.com/homedetails/15949-Arena-Dr-Ramona-CA-92065/16754450_zpid/"]
# for link in links:

for link in df['links']:
  scores_page = s.get(link, headers=req_headers)
  soup = BeautifulSoup(scores_page.content, 'html.parser')
  spans = soup.find_all("meta",  property="zillow_fb:description")
  sold_price=apply_regex_sold(str(spans))
  sold_prices.append(sold_price)

print(sold_prices)
```

```
cols=['Sold_Price']
Z_sold_price = pd.DataFrame(sold_prices, columns=cols)
```
